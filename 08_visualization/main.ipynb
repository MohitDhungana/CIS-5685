```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("./Store11_clean.csv")
df.head()
```

##### Scatter plot Sales and Labor_Cost


```python
sns.scatterplot(data=df, x="Labor_Cost", y="Sales")
plt.show()
```

##### Line plot Date and Item_Labor


```python
sns.lineplot(data=df, x="Date", y="Item_Labor")
plt.show()
```

##### Create a histogram for a column of your choice


```python
sns.histplot(df, x="Bad_Weather_Days")
```

##### Create a bar chart using Bad_Weather_Days and Sales


```python
sns.barplot(data=df, x="Bad_Weather_Days", y="Sales")
plt.show()
```
