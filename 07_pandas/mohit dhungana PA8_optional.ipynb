```python
import pandas as pd
```

```python
boone_df = pd.read_csv("boone_clean.csv")
boone_df
```

```python
boone_df["month"] = pd.to_datetime(boone_df["Datetime"]).dt.month
boone_df["day"] = pd.to_datetime(boone_df["Datetime"]).dt.day
boone_df["hour"] = pd.to_datetime(boone_df["Datetime"]).dt.hour
boone_df
```

```python
test = boone_df[
    (boone_df["month"] == 1) & (boone_df["day"] == 1) & (boone_df["hour"] == 0)
]
test
```

```python
mean_df = boone_df.groupby(["month", "day", "hour"]).agg(
    {
        "AirTemp": "mean",
        "DewPoint": "mean",
        "WindDir": "mean",
        "WindSpd": "mean",
        "SkyCond": "mean",
        "Precip1Hr": "mean",
    }
)
```

```python
mean_df.to_csv("mean.csv")
```
